```python
#Ex0.修改前
import numpy as np
import matplotlib.pyplot as plt
def y(th,x):
    return np.tan(th)*x-g/(2*v**2*np.cos(th)**2)*x**2
v=8 ; g=9.8
th = 30*np.pi/180
x=np.linspace(0,v**2/g*np.sin(2*th),1001)
plt.plot(x,y(30*np.pi/180,x))
x=np.linspace(0,v**2/g*np.sin(2*th),1001)
plt.plot(x,y(45*np.pi/180,x),color='red')
x=np.linspace(0,v**2/g*np.sin(2*th),1001)
plt.plot(x,y(60*np.pi/180,x),color='green')
plt.grid('True')
plt.show()
```

```python
#Ex0.修改後
import numpy as np
import matplotlib.pyplot as plt
def projectile(th,x):
    return np.tan(th)*x-g/(2*v**2*np.cos(th)**2)*x**2
# parameters
v=8                 # Initial velocity ( m/s )
g=9.8              # Gravity ( m/s² )
angles = [15, 30, 45, 60, 75]
colors = ['purple','blue','red','green','yellow']
plt.figure(figsize=(10,6))              #設定圖的大小，單位英寸（inch）; 寬10英吋，高6英吋
plt.scatter( 0, 0, color='orange', s= 40, label= "Launch Point")                 #Launch Point
for angle,color in zip(angles , colors):
    th = np.deg2rad(angle)
    xmax = v**2/g*np.sin(2*th)
    x = np.linspace(0,xmax,1001)
    plt.plot (x,  projectile(th,x), color = color, label=f"{angle} degrees" )      #圖例
plt.title("Projectile Motion Simulation")                                                       #圖標題
plt.xlabel("Horizontal Distance (m)")                                                           #x坐標軸
plt.ylabel("Height (m)")                                                                               #y坐標軸
plt.legend()
plt.grid(True)
plt.savefig("projectile_motion.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#ex1.a修改後
import matplotlib.pyplot as plt
import numpy as np
T = 10                                              #運動週期
w = 2*np.pi/T                                  #角動量
t = np.linspace(0,10,200)               #時間軸向量
x_trajectory = 2*np.cos(w*t)           #x軸振幅
y_trajectory = np.sin(w*t)               #y軸振幅
t0 = 0.5
p = np.array([2*np.cos(w*t0), np.sin(w*t0)])
v = np.array([-w*2*np.sin(w*t0), w*np.cos(w*t0)])
a = np.array([-(w**2)*2*np.cos(w*t0), -(w**2)*np.sin(w*t0)])
plt.figure( figsize=(10,10), dpi=300)               #設定高解析度正方形畫布
plt.plot(x_trajectory, y_trajectory, color="Green",linestyle="--", label="Trajectory")
plt.plot(0,0,'o', label="Origin (Center)", color="pink")
plt.plot(p[0],p[1], 'ro', markersize=10, label=f"Particle at t={t0}s")      #繪製質點位置（P）
#繪製向量箭頭
plt.arrow(p[0],p[1],v[0],v[1], head_width=0.05, head_length=0.05, color='blue', label="Velocity (v)")
plt.arrow(p[0],p[1],a[0],a[1], head_width=0.05, head_length=0.05, color='purple', label="Acceleration (a)")
plt.axis('equal')                                                #確保XY軸比例為1:1
plt.grid(True, linestyle=":", alpha=0.6)
plt.title(f"Kinematics Analysis on Elliptical Orbit (t = {t0}s)")
plt.xlabel("x Position", fontsize=10)
plt.ylabel("y Position", fontsize=10)
plt.legend(loc="upper right")                            #避免重複標籤
#圖片儲存
plt.savefig("kinematics_ellipse.png", dpi=300, bbox_inches="tight")
plt.show()
```
